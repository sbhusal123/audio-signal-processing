## 1.1 What is a Signal?

- A **signal** is anything that carries information and changes over time.

- In audio, a signal is usually a **pressure wave** that travels through air and is captured as a voltage by microphones.

**Example:**

Your voice speaking into a mic creates a pressure wave â†’ mic converts it to a voltage signal â†’ computer stores it.


## 1.2 What is an Audio Signal?

- An **audio signal** is a representation of sound, typically as a function of time.

- **Analog audio:** A continuous wave (e.g., real-world sounds).

- **Digital audio:** A sequence of numbers (discrete samples) that approximate the wave.

ðŸ‘‰ In computers, we work with **digital audio** â€” a list of amplitude values over time.


## 1.3 Time Domain vs Frequency Domain

- **Time Domain:**
    You look at how the signal changes over time (Amplitude vs Time).

- **Frequency Domain:**
    You analyze which frequencies are present in the signal (Amplitude vs Frequency).

**Summary:**

- Time domain tells when something happened.

- Frequency domain tells what tones (pitches) were present.

## 1.4 Basics of Periodic Signals:

- **Periodic signals** repeat over time.

- The simplest example: **Sine Wave.**

**Sine Wave Formula:**

$x(t) = Asin(2\pi f t+\phi)$

where:
- **A** = amplitude (loudness)
- **$\phi$** = frequency (pitch in Hz)
- **t** = time

## ðŸ“Œ Exercise 1: Plot a Sine Wave

```python
import numpy as np
import matplotlib.pyplot as plt

# Settings
frequency = 440  # Frequency in Hz (A4 note)
sampling_rate = 44100  # Samples per second
duration = 1.0  # seconds

# Time axis
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

# Sine wave
amplitude = 0.5
y = amplitude * np.sin(2 * np.pi * frequency * t)

# Plot
plt.plot(t[:1000], y[:1000])  # Plot only first 1000 samples
plt.title("Sine Wave - 440 Hz")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

## ðŸ“Œ Exercise 2: Plot a Square and Sawtooth Wave

```python
from scipy import signal

# Square wave
square_wave = signal.square(2 * np.pi * frequency * t)

# Sawtooth wave
sawtooth_wave = signal.sawtooth(2 * np.pi * frequency * t)

# Plot both
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(t[:1000], square_wave[:1000])
plt.title("Square Wave")

plt.subplot(1, 2, 2)
plt.plot(t[:1000], sawtooth_wave[:1000])
plt.title("Sawtooth Wave")

plt.show()
```
